The dataset contains 1465 records and 16 columns related to Amazon products, including pricing, discounts, customer ratings, and reviews. While the dataset is complete overall, most numeric fields such as prices, ratings, and discount percentages are stored as text, indicating the need for data cleaning before analysis.

```python
df['discounted_price'] = (
    df['discounted_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
df['actual_price'] = (
    df['actual_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%', '', regex=False)
    .astype(float)
)
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
df['rating_count'] = df['rating_count'].fillna(0)
```

## Exploratory Data Analysis

- Most products have ratings above 4, indicating generally high customer satisfaction.
- Discounted prices show a wide range, with a few high-priced outliers.
- Discount percentages vary significantly, with many products offering medium to high discounts.
- Initial exploration shows no strong relationship between higher discounts and higher ratings.


```python
price_bins = [0, 500, 1000, 2000, 5000, 10000, 100000]

df.groupby(
    pd.cut(df['discounted_price'], bins=price_bins),observed=True
)['rating'].mean()
```

```python
df.groupby(
    pd.cut(df['discount_percentage'], bins=[0,20,40,60,80,100]),observed=True
)['rating'].mean()
```

```python
category_rating = (
    df.groupby('category')['rating']
    .mean()
    .sort_values(ascending=False)
)

category_rating.head(10)
```

```python
high_discount = df[df['discount_percentage'] > 60]['rating'].mean()
low_discount = df[df['discount_percentage'] <= 60]['rating'].mean()

high_discount, low_discount
```

## Analysis & Business Insights

- Products priced in the mid-range tend to receive higher average ratings compared to very low or very high-priced products.
- Moderate discounts are associated with better customer ratings, while very high discounts show slightly lower satisfaction.
- Categories such as Computers & Accessories (Tablets) perform better in terms of customer ratings.
- Products with extremely high discounts generally have lower average ratings, indicating that discounts alone do not guarantee customer satisfaction.


## Final Insights

- Overall customer satisfaction is high, with most products rated above 4.0.
- Discounted prices vary widely, showing that the dataset includes both low-cost and premium products.
- Actual prices are consistently higher than discounted prices, indicating that discounts play an important role in pricing strategy.
- Products with very high discounts tend to have slightly lower ratings, suggesting that higher discounts do not necessarily improve customer satisfaction.
- Certain categories, such as Computers & Accessories (Tablets), show higher average ratings compared to others.

## Conclusion

This project analyzed Amazon product data using NumPy and Pandas to understand customer satisfaction and pricing patterns. The analysis shows that product quality and category have a stronger influence on ratings than heavy discounts. These insights can help businesses focus on product improvement and strategic pricing rather than relying solely on high discounts.
